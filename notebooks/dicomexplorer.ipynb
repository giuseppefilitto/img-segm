# Dicom Explorer

```python
import sys
sys.path.append('../')

from MRIsegm.utils import get_slice, get_slice_info, get_rois, explore_slice, explore_roi, explore_mask, explore_applied_mask
from ipywidgets import interact, fixed
```

## Getting Slice

```python
src = '/Users/giuseppefilitto/Pazienti_anonym_sorted/'
patience = 'BO11/'
dir_path =  src + patience + 'T2AX'

slice = get_slice(dir_path=dir_path) 

get_slice_info(slice=slice)
```

## Getting ROIs

```python
roi_path = src + patience + 'T2ROI'

positions, xs, ys = get_rois(roi_path=roi_path)       
```

## Plotting

### Slice

```python
layer = 14
explore_slice(slice = slice, layer=layer)
```

### ROI

```python
explore_roi(slice=slice, layer=layer, positions=positions, xs=xs, ys=ys)
```

### Mask

```python
explore_mask(slice=slice, layer=layer, positions=positions, xs=xs, ys=ys)
```

### Appling mask

```python
explore_applied_mask(slice=slice, layer=layer, positions=positions, xs=xs, ys=ys)
```

## Interactive Exploration

```python
interact(explore_slice, slice=fixed(slice), layer=(0, slice.shape[0]-1) )
```

```python
interact(explore_roi, slice=fixed(slice), layer=(positions[0], positions[-1]), positions=fixed(positions), xs=fixed(xs), ys=fixed(ys))
```

```python
interact(explore_applied_mask, slice=fixed(slice), layer=(positions[0], positions[-1]), positions=fixed(positions), xs=fixed(xs), ys=fixed(ys))
```
